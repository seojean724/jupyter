# 시간 측정하기

- 시간을 측정해서 더 나은 방법을 찾는다.
- 먼저 작은 데이터에 대해서 테스트를 진행하고 큰 데이터를 적용해 보자.

```python
import time

start_time = time.time()
count = 0
for i in range(100000):
    count += 1
print count

end_time = time.time()

print "duration: ", end_time - start_time
```

```python
products = gl.SFrame('../data/amazon_baby.gl/')
products['word_count'] = gl.text_analytics.count_words(products['review'])
filtered_products = products[products['rating'] != 3]

selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']

def selected_word_count(dict, w):
    if w in dict:
        return dict[w]
    else:
        return 0
```

### loop로 직접 세기 BOMSHIK's solution

```python
start_time = time.time()

temp_wc = filtered_products['word_count']
temp_len = len (temp_wc)

for w in selected_words:
    i = 0 
    cnt = 0
    while i < temp_len :
        cnt = cnt + selected_word_count(temp_wc[i], w)
        i = i+1
    print w + " : " + str(cnt)
    
end_time = time.time()
print "duration: ", end_time - start_time
```

이 솔루션은 369.4초가 걸렸습니다.

### lambda와 apply 사용하기

```python
start_time = time.time()

for w in selected_words:
    filtered_products[w] = filtered_products['word_count'].apply(lambda x:selected_word_count(x, w))
    print w, filtered_products[w].sum()
    
end_time = time.time()
print "duration: ", end_time - start_time
```

이 솔루션은 55.6초가 걸렸습니다.

### 결론
 
 - length를 계산하는 게 불필요하다.
 - apply의 강력함을 알 수 있다.
 - 카운트를 개별적으로 세는 것보다 SArray의 sum으로 계산하는 게 훨씬 빠르다.
 
**데이터가 많을수록 효과는 극적으로 벌어진다. 따라서 apply와 sum을 잘 활용하자.**

# pandas의 get_dummies

카테고리 값이 2개 이상인 feature들을 각각의 개별 컬럼으로 만들어 줍시다. 해당 결과를 **X_dummies**에 저장합시다.

    pd.get_dummies(데이터셋, 변환하고자하는컬럼이름들리스트, drop_first=True)
    

```python
full_df = pd.concat([train_df.drop('Survived', axis=1), test_df], axis=0)
```

```python
full_df.head()
```

```python
converting_columns = ['Pclass']
drop_features = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Sex', 'Age']
features = list(set(full_df.columns.values.tolist()) - set(drop_features))
```

```python
X_dummies = pd.get_dummies(full_df[features], 
                           columns = converting_columns,
                           drop_first=True)
```

```python
X_dummies.head()
```

첫번째 이미지의 R 색상의 맨 윗줄의 데이터가 출력됩니다.

```python
def normalize_pixel(array):
    new_list = []
    for v in array:
        new_list.append(v/255)
    return new_list
```

```python
image_train_norm = image_train['image_array'][0:1].apply(normalize_pixel)
```
